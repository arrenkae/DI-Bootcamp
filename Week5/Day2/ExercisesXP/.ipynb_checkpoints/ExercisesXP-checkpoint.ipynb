1. Start by importing necessary libraries and create your own dataset using the pd.date_range function. Assume the dataset has the following columns: [‘Date’, ‘Temperature’, ‘Humidity’, ‘Wind_Speed’]. You can use the following code to create the data points for Temperature, Humidity and Wind_Speed columns:
```
 temperature = [30, 31, 29, 32, 33, 34, 28, 27, 25, 35]
 humidity = [80, 75, 70, 85, 90, 80, 75, 60, 65, 70]
 wind_speed = [5, 6, 7, 8, 9, 4, 10, 5, 6, 7]

 # Create the Date column here (with dates between 2021-01-01 and 2021-01-10, Day frequency)
 date_rng = ...

 # Create a dataframe with all of the columns inside
 weather_data = pd.DataFrame(...)
```
2. Determine the frequency of the dataset.
- Hint: Frequency refers to the regularity with which the data points are recorded, e.g., daily, weekly, etc. Use pd.infer_freq() to infer the frequency of your DataFrame’s ‘Date’ column.
*The output should look like this:*
> The frequency of the dataset is D.

3. Set ‘Date’ as the index of the DataFrame.
- Hint: You can use the set_index method to set ‘Date’ as the index of the DataFrame. Use the inplace=True parameter to modify the DataFrame in place.

````python
date_rng = pd.date_range(start='2021-01-01', end='2021-01-10', freq='D')
date_rng
````

````python
weather_data = pd.DataFrame({ 
    'temperature':[30, 31, 29, 32, 33, 34, 28, 27, 25, 35],
    'humidity':[80, 75, 70, 85, 90, 80, 75, 60, 65, 70],
    'wind_speed':[5, 6, 7, 8, 9, 4, 10, 5, 6, 7],
    'date':date_rng
})
weather_data
````

````python
print(f"The frequency of the dataset is {pd.infer_freq(weather_data['date'])}")
````

````python
weather_data.set_index('date', inplace=True)
weather_data
````
