# Part I
1. **Load the Data:** Import the California Housing dataset from the sklearn library and convert it into a pandas DataFrame.

```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['Target'] = data['target']
```

3. **Preprocess the Data:**
- Check for missing data using df.isnull().sum().
- Normalize continuous variables using MinMaxScaler from sklearn.

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[df.columns]), columns=df.columns)
df_scaled
```

4. **Categorical Variables:**
- Simulate a categorical variable by categorizing the ‘HouseAge’ column.
- Hint: Use pd.cut() to create age categories like ‘New’, ‘Old’, and ‘Very Old’ and then apply one-hot encoding.

```python
age = pd.cut(df['HouseAge'], 3, labels=['New', 'Old', 'Very Old'])
age
```

```python
encoded = pd.get_dummies(age, dtype=int)
encoded
```

```python
df = pd.concat([df, encoded], axis=1)
df
```

5. **Correlations:**
- Create a correlation matrix using df.corr().
- Hint: For visualization, use matplotlib. Example: plt.matshow().

```python
corr = df.corr()
corr
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr, cmap='coolwarm');
fig.colorbar(cax);
ax.set_xticks(np.arange(len(corr)))
ax.set_yticks(np.arange(len(corr)))
ax.set_xticklabels(corr.index, rotation=90);
ax.set_yticklabels(corr.index);
```

6. **Feature Engineering:**
- Create a new feature derived from existing columns.
- Hint: You can combine ‘MedInc’ and ‘AveRooms’.

```python
df['IncRooms'] = df['MedInc'] / df['AveRooms']
df.head()
```

# Part II
1. Analyse the relationship between Property Age and Price

```python
age_price = encoded.set_index(df['Target'].astype('int'))
age_price = age_price.groupby('Target').sum()
age_price.plot(
    kind='bar',
    rot = 0,
    ylabel='Frequency',
    title='Property Age vs. Median Value'
);
```

# Part III
1. Analyse the Neighborhood Crime Rate Impact on Prices

```python
income_bins = pd.cut(df['MedInc'], 5, labels=['0-2', '2-4', '4-6', '6-8', '8+'])
income_bins
```

```python
target_by_income = df.groupby(income_bins, observed=False)['Target'].mean()
target_by_income
```

```python
target_by_income.plot(
    grid=True,
    style='.-',
    xlabel='Income Bin',
    ylabel='Average Target',
    title='Impact of Neighborhood Median Income on Prices'
);
```
