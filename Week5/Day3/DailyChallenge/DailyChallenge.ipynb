# Task 1:

1. Import the numpy, pandas, and matplotlib.pyplot packages.

2. Use numpy to create an array x of 100 numbers ranging from 0 to 20 (inclusive) using the linspace function.

3. Create an array y which is the square of x.

4. Use matplotlib to plot x vs y.

5. Your plot should have an appropriate title and labels for the x and y axes. You should also adjust the x and y limits to correctly show all the data.

6. Save this plot as a .png file using savefig.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 100)
y = np.square(x)
plt.title('Plot of x vs y (y=x^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0.0, 20.0)
plt.ylim(0, 400)
plt.plot(x,y);
plt.savefig('task1.png')
```

# Task 2:
1. Import the dataset from this link using pandas and assign it to a variable ‘chipo’.

2. Clean the item_price column and transform it in a float. Remove the dollar sign ($) from the ‘item_price’ column using the apply function in combination with a lambda function. Then convert the ‘item_price’ column to the float datatype.

3. Create a scatter plot presenting the relationship between the price of items and the quantity ordered. Use the ‘item_price’ column for the x-axis and the ‘quantity’ column for the y-axis.

4. Your plot should have an appropriate title and labels for the x and y axes.

5. Save this plot as a .png file using savefig.

```python
chipo = pd.read_table('../../Day2/ExercisesXP/chipotle.tsv')
chipo.head()
```

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.strip('$')))
chipo.head()
```

```python
chipo.plot.scatter(
    x='item_price',
    y='quantity',
    title='Relationships between Item Price and Quantity Ordered',
    xlabel = 'Item Price',
    ylabel = 'Quantity Ordered'
)
plt.savefig('task2.png')
```

# Task 3:

1. Group the ‘chipo’ data by item_name and calculate the total quantity for each item.

2. Create a bar plot that presents the total quantities for the top 5 items. Use the ‘item_name’ column for the x-axis and the calculated total quantity for each item for the y-axis.

3. Your plot should have an appropriate title and labels for the x and y axes.

4. Save this plot as a .png file using savefig.

```python
chipo.groupby('item_name')['quantity'].sum()
```

```python
chipo.groupby('item_name')['quantity'].sum().nlargest(5).plot(
    kind='bar',
    rot=45,
    width=0.8,
    title='Top 5 Items by Quantity Ordered',
    xlabel = 'Item Name',
    ylabel = 'Total Quantity Ordered'
);
plt.savefig('task3.png')
```
