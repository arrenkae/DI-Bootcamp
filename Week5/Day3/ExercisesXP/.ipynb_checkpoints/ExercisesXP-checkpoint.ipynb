
For exercises 7-9: You can use the following data for producing the plots:

````python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
````

# Exercise 7 :
1. Create a line plot and scatter plot representing your favorite mathematical functions. Change the line color to tab:blue for the line plot and use a marker of your choice for the scatter plot.

````python
plt.title('y = sin(x)')
plt.plot(x,y1, color='tab:blue')
````

````python
plt.title('y = cos(x)')
plt.scatter(x,y2, color='tab:blue', marker='o')
````

# Exercise 8 :
1. Plot the line plot and scatter plot from Exercise 7 on the same figure using subplots. Use plt.subplots to create a figure with two subplots.

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(x, y1, color='tab:blue')
ax2.scatter(x, y2, color='tab:blue', marker='o')
plt.show()
````

# Exercise 9 :
1. Modify the plots from Exercise 8 to include a legend. Use the legend function from Matplotlib.

````python
fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(x, y1, color='tab:blue', label='y = sin(x)')
ax1.legend()
ax2.scatter(x, y2, color='tab:blue', marker='o', label='y = cos(x)')
ax2.legend()
plt.show()
````
