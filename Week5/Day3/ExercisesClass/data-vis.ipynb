# Assignment 1: Simple Line Chart

Build a simple line plot for the oil series. Are there any interesting patterns here? 

What happened in 2014 to oil price?

```python
oil = pd.read_csv('./oil.csv')
oil
```

```python
oil.plot(x='date');
```

```python
(oil
.astype({'date':'datetime64[ns]'})
.set_index('date')
.loc['2013':'2014']
.plot())
```

```python
oil.plot(
    x='date',
    title='oil transactions'
)
```

```python
trans_sm.plot(
    x='date',
    title='Transactions',
    ylabel='amount',
    color=['pink', 'red']
)
```

```python
trans_sm.plot(
    x='date',
    title='Transactions',
    ylabel='amount',
    cmap='Pastel2'
)
```

```python
trans_sm.plot(
    x='date',
    title='Transactions',
    ylabel='amount',
    cmap='Pastel2',
    style=['--','-.']
).legend(loc='center left')
```

```python
trans_sm.plot(
    kind='bar',
    rot=45
)
```

```python
trans_sm.plot.pie(y='transactions')
```

```python
trans_sm.plot.scatter(y='transactions',x='date')
```
